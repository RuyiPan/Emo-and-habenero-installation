```r
keywords <- emo::ji_keyword
keywords <- keywords[lengths(keywords) > 1]
emojis <- purrr::map_chr(keywords, function(x) paste0(emo::ji_name[x], collapse = ""))
cat(paste(names(emojis), ": ", emojis, collapse = "\n"))
```
